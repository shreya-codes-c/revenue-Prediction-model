## new model

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
```

```python
import math
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python

model_1=Lasso(max_iter=500,tol=0.01,random_state=42)
model_2=Ridge(max_iter=500,tol=0.01,random_state=42)
model_3=RandomForestRegressor(n_estimators=500,max_depth=5,min_samples_split=4,random_state=42)
model_4=RandomForestRegressor(n_estimators=100,max_depth=7,random_state=42)
model_5=GradientBoostingRegressor(n_estimators=500,max_depth=4,learning_rate=0.1)
model_6=LinearRegression()
models=[model_1,model_2,model_3,model_4,model_5,model_6]
```

```python
def score_model(model,Xtrain=X_train,Xtest=X_test,ytrain=y_train,ytest=y_test):
    model.fit(Xtrain,ytrain)
    preds=model.predict(Xtest)
    return mean_squared_error(ytest,preds)
for i in range (0,len(models)):
    mse=score_model(models[i])
    rmse=math.sqrt(mse)
    print("model %d RMSE: %d" %(i+1,rmse))
```
